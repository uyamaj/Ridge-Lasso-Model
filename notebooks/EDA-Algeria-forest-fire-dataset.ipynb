```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Data Cleaning

```python
## missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is divided into two subsets based on the Region starting from the 122nd index. We can create a new column to indicate the Region:

1. "Bejaia Region Dataset"
2. "Sidi-Bel Abbes Region Dataset"

Add a new column with the Region name.


```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.head()
```

```python
df.info()
```

```python
# convert the region column to int
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.iloc[[122]]
```

```python
##drop the 122nd row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in column names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
## Convert the object datatypes to integers
df.columns 
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

#### Convert the other columns to float data datatype

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
## Save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

##  Exploratory Data Analysis

```python
## drop day,month and year columns
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
## categories in classes
df_copy['Classes'].value_counts()
```

```python
## Encoding of the categories in classes
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
## Plot desnity plot for all features
plt.style.use('seaborn')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## Percentage for Pie Chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# plotting piechart
classlabels=["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
## Monthly Fire Analysis
dftemp=df.loc[df['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Fire Analysis of Sidi- Bel Regions",weight='bold')
```

```python
## Monthly Fire Analysis
dftemp=df.loc[df['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Fire Analysis of Brjaia Regions",weight='bold')
```

It is observed that August and September had the highest number of forest fires in both regions. From the above plot of monthly data, we can deduce a few things:

The majority of the fires occurred in August, with a significant spike in fire activity concentrated in just three months: June, July, and August.
September experienced fewer fires compared to the peak months.
